```python
import pandas
pandas.options.mode.chained_assignment = None  # default='warn'

purchase_data = pandas.read_csv(r'C:\Users\Mason\nu-chi-data-pt-09-2020-u-c\04-Pandas\HW\Instructions\HeroesOfPymoli\Resources\purchase_data.csv')

PurchaseDF = pandas.DataFrame(purchase_data)
PurchaseDF.head()
```

```python
#Player count
Player_Count = len(PurchaseDF['SN'].drop_duplicates())
PlayerCountDF = pandas.DataFrame([Player_Count], columns=['Player Count'])
PlayerCountDF = PlayerCountDF.style.hide_index()
PlayerCountDF
```

```python
#Number of unique items
Unique_Items = len(PurchaseDF['Item Name'].drop_duplicates())

#Average purchase price
Average_Price = round(PurchaseDF['Price'].mean(), 2)

#Total number of purchases
Total_Purchases = len(PurchaseDF.index)

#Total revenue
Total_Revenue = round(PurchaseDF['Price'].sum(), 2)

PurchasingAnalysis = {'Unique Items': [Unique_Items], 'Average Price': [Average_Price], 'Total Purchases': [Total_Purchases], 'Total Revenue': [Total_Revenue]}
PurchasingAnalysisDF = pandas.DataFrame(data=PurchasingAnalysis)
PurchasingAnalysisDF = PurchasingAnalysisDF.style.hide_index()
PurchasingAnalysisDF = PurchasingAnalysisDF.format({'Average Price': '${:20,.2f}', 'Total Revenue': '${:20,.2f}'})
PurchasingAnalysisDF
```

```python
#Unique player count
UniqueDF = PurchaseDF.drop_duplicates(subset='SN')

#Unique player count by gender
GenderCount = UniqueDF['Gender'].value_counts()

#Percent of players by gender
GenderPercentage = UniqueDF['Gender'].value_counts(normalize=True) * 100

GenderDF = pandas.DataFrame({'Total Count': GenderCount, 'Percentage of Players': GenderPercentage})
GenderDF = GenderDF.round(2)
GenderDF
```

```python
GPurchaseDF = PurchaseDF.groupby('Gender').agg({'Purchase ID': 'count', 'Price': ['mean', 'sum']})

PerPersonAveragePurchaseG = GPurchaseDF['Price']['sum']/GenderDF['Total Count']
GPurchaseDF['Average Total Purchase Per Person'] = pandas.Series(PerPersonAveragePurchaseG)

GPurchaseDF = GPurchaseDF.rename(columns = {'count': 'Purchase Count', 'mean': 'Average Purchase Price', 'sum': 'Total Purchase Value'})
GPurchaseDF = GPurchaseDF.round(2)
GPurchaseDF
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
bin_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '>40']

PurchaseDF['Age Range'] = pandas.cut(PurchaseDF['Age'], bins, labels = bin_names, include_lowest = True)

APurchaseDF = PurchaseDF.groupby('Age Range').agg({'Purchase ID': 'count', 'Price': ['mean', 'sum']})

APurchaseDF = APurchaseDF.round(2)
APurchaseDF = APurchaseDF.rename(columns = {'count': 'Purchase Count', 'mean': 'Average Purchase Price', 'sum': 'Total Purchase Value'})
APurchaseDF
```

```python
SNDF = PurchaseDF.groupby('SN')

SpenderPC = SNDF['Purchase ID'].count()
SpenderAPP = SNDF['Price'].mean()
SpenderTPV = SNDF['Price'].sum()

TopSpendersDF = pandas.DataFrame({'Purchase Count': SpenderPC, 'Average Purchase Price': SpenderAPP, 'Total Purchase Value': SpenderTPV})
TopSpendersDF = TopSpendersDF.sort_values('Total Purchase Value', ascending = False)
TopSpendersDF = TopSpendersDF.round(2)
TopSpendersDF.head()
```

```python
ItemDF = PurchaseDF.groupby(['Item ID', 'Item Name'])

PopularPC = ItemDF['Purchase ID'].count()
PopularAPP = ItemDF['Price'].mean()
PopularTPV = ItemDF['Price'].sum()

PopularDF = pandas.DataFrame({'Purchase Count': PopularPC, 'Average Purchase Price': PopularAPP, 'Total Purchase Value': PopularTPV})
PopularDF = PopularDF.sort_values('Purchase Count', ascending = False)
PopularDF = PopularDF.round(2)
PopularDF.head()
```
